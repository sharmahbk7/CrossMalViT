# Training Demo

Run a quick forward/backward pass on synthetic data.

```python
import torch
from crossmal_vit.models import CrossMalViT

model = CrossMalViT(pretrained=False)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)

batch = {
    'raw': torch.randn(2, 1, 224, 224),
    'entropy': torch.randn(2, 1, 224, 224),
    'frequency': torch.randn(2, 1, 224, 224),
}
labels = torch.tensor([0, 1])

outputs = model(batch)
loss = torch.nn.functional.cross_entropy(outputs['logits'], labels)
loss.backward()
optimizer.step()
print('Loss:', float(loss))
```
