# Multi-View Generation

Generate raw, entropy, and frequency views.

```python
from pathlib import Path
import numpy as np
from PIL import Image
from crossmal_vit.data.transforms import MultiViewTransform

sample_path = Path('../data/sample.png')
if sample_path.exists():
    image = Image.open(sample_path).convert('L')
    transform = MultiViewTransform(img_size=224, entropy_window=9)
    views = transform(image)
    for name, tensor in views.items():
        print(name, tensor.shape, tensor.min().item(), tensor.max().item())
else:
    print('Sample image not found at ../data/sample.png')
```
